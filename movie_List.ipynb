```python
#@title Get movie list
URL = "https://katmoviehd.rs/category/romance/" #@param {type:"string"}
#@markdown Input katmoviehd and extramovies website url only.
SHOW_ITEM = 10 #@param {type:"integer"}
SORT_DEC = True #@param {type:"boolean"}
from lxml import html
from bs4 import BeautifulSoup as bs
from IPython.display import clear_output
import re
import requests
listal=[]

def Extralst(u):
  global rate
  global inp
  r=requests.get(u)
  sop=bs(r.content,'lxml')
  for m in sop.find_all('div', class_='imag'):
    name = m.find('h2').a.text
    rate = m.find("span", class_="imdb").text
    link = m.a['href']
    listal.append([rate, name, link])
    printAsSorted(listal)
 
def Katlst(u):
  global rate
  global inp
  global link
  r=requests.get(u)
  sop=bs(r.content,'lxml')
  for m in sop.find_all('div', class_='post-content'):
    u = m.h2.a['href']
    r=requests.get(u)
    inp=bs(r.content,'lxml')
    try:
      name = inp.find('h1', class_='entry-title').a.text
      try:
        rate = re.search(r"[>|\s](\d\.\d)[\/10|<|\n]", inp.text, re.M).group(1)
      except:
        rate = 0.0
      link = u
      # print (name[:25],'\t:',rate,'\t', link)
      listal.append([rate, str(name), link])
    except:
      pass
    printAsSorted(listal)

def printAsSorted(listal):
  clear_output(True)
  for i,p in enumerate(sorted(listal, key=lambda p:str(p[0]), reverse=SORT_DEC)):
    print(p[1][:25],'\t:',p[0],'\t', p[2])
    if i==SHOW_ITEM:
      break
    

def main(l):
  global u
  
  r=requests.get(l)
  if "katmoviehd" in l:
    t = 0
    rg = bs(r.content, 'lxml')
    rg = rg.find('div', class_='navigation')
    rg = rg.find_all('a', class_='page-numbers')[-2].text
  elif "extramovies" in l:
    t = 1
    tree = html.fromstring(r.content)
    dd = tree.xpath('//span[@class="pages"]/text()')[0]
    rg = dd.split('of')[1]
  rg = int(rg)
  for u in range(1, rg):
    if t==0:
      Katlst(l+f"/page/{u}/")
    elif t==1:
      Extralst(l+f"/page/{u}/")
    print(f'Completed: {u+1}/{rg} | Total : {len(listal)}')
    

main(URL)
```

## Save as Csv

```python
import pandas
from google.colab import files

rate = []
name = []
link = []
for l in listal:
  rate.append(l[0])
  name.append(l[1])
  link.append(l[2])

df = pandas.DataFrame(data={"name": name, "rate": rate, "link": link})
df.to_csv("df.csv", sep=',', index=False)
files.download('df.csv')
```
