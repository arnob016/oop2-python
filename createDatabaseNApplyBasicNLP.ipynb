```python
## lower casing 
df['description'] = df['description'].apply(lambda x: x.lower())
df
```

```python
import re

## remove html tags
def removeHtml(data):
  pattern = re.compile(r'<.*?>')
  return pattern.sub('', data)

df['description'] = df['description'].apply(removeHtml)
df
```

```python
## remove url
def removeUrl(data):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub('', data)

df['description'] = df['description'].apply(removeUrl)
df
```

```python
import string
## remove puncuation
def removePunc(data):
  return data.translate(str.maketrans('', '', string.punctuation))

df['description'] = df['description'].apply(removePunc)
df
```

```python
## spelling correction
from textblob import TextBlob
def spC(data):
  textBlobIns = TextBlob(data)
  return textBlobIns.correct().string

df['description'] = df['description'].apply(spC)
df
```

```python
## ready stopwords nltk lib
import nltk
nltk.download('stopwords')
```

```python
## remove stop words
from nltk.corpus import stopwords

def removeStopwords(text):
    new_text = []
    
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)

df['description'] = df['description'].apply(removeStopwords)
df
```

```python
### NLTK
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

df['description'] = df['description'].apply(word_tokenize)
df
```
